```python
import os
import time
import torch
import numpy as np
import pandas as pd
from tqdm import tqdm
import pickle
import faiss

# Importações do LangChain
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.schema.document import Document
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS

# Importação corrigida para RetrievalQA
from langchain.chains import RetrievalQA
# Outras importações podem ser necessárias ou não, dependendo do uso
# from langchain_community.retrievers import ContextualCompressionRetriever
# from langchain_community.llms import HuggingFaceHub
# from langchain_openai import ChatOpenAI  # Se estiver usando OpenAI
from langchain.prompts import PromptTemplate

# Definir o dispositivo (GPU ou CPU)
device = "cuda" if torch.cuda.is_available() else "cpu"
print("Usando dispositivo:", device)

# Modelo de embeddings do Hugging Face
model_pre = "intfloat/multilingual-e5-base"
print(f"Usando modelo de embeddings: {model_pre}")
```

```python
class CGURecommendationPipeline:
    """
    Pipeline de recomendação para recursos de acesso à informação da CGU
    utilizando LangChain para processamento e recuperação.
    """

    def __init__(self, embedding_model_name, device="cpu"):
        """
        Inicializa o pipeline com o modelo de embeddings especificado.

        Args:
            embedding_model_name: Nome do modelo de embeddings do HuggingFace
            device: Dispositivo para processamento ("cpu" ou "cuda")
        """
        self.embedding_model_name = embedding_model_name
        self.device = device

        # Inicializar modelo de embeddings
        self.embeddings = HuggingFaceEmbeddings(
            model_name=embedding_model_name,
            model_kwargs={"device": device}
        )

        # Armazenar vectorstores
        self.vectorstore_pedidos = None
        self.vectorstore_recursos = None

    def save_embeddings_to_pickle(self, ids, embeddings, file_path, include_metadata=None):
        """
        Salva embeddings em formato pickle para carregamento rápido.

        Args:
            ids: Lista de IDs (ProtocoloPedido ou IdRecurso)
            embeddings: Array numpy com os embeddings
            file_path: Caminho completo para salvar o arquivo pickle
            include_metadata: DataFrame com metadados a incluir (opcional)
        """
        # Criar dicionário com IDs e embeddings
        embeddings_dict = {
            'ids': ids,
            'embeddings': embeddings,
            'model_name': self.embedding_model_name,
            'created_at': time.time()
        }

        # Adicionar metadados se fornecidos
        if include_metadata is not None:
            # Converter para um formato mais leve
            metadata_dict = include_metadata.to_dict(orient='records')
            embeddings_dict['metadata'] = metadata_dict

        # Salvar em pickle
        with open(file_path, 'wb') as f:
            pickle.dump(embeddings_dict, f)

        print(f"Embeddings salvos em {file_path}")
        print(f"Total de {len(ids)} embeddings salvos, tamanho do modelo: {embeddings.shape}")
        return file_path

    def load_embeddings_from_pickle(self, file_path):
        """
        Carrega embeddings de arquivo pickle.

        Args:
            file_path: Caminho para o arquivo pickle

        Returns:
            Tupla (ids, embeddings, metadata)
        """
        print(f"Carregando embeddings de {file_path}...")
        start_time = time.time()

        try:
            with open(file_path, 'rb') as f:
                data = pickle.load(f)

            ids = data['ids']
            embeddings = data['embeddings']
            metadata = data.get('metadata', None)
            model_name = data.get('model_name', 'unknown')

            print(f"Embeddings carregados em {time.time() - start_time:.2f} segundos")
            print(f"Modelo original: {model_name}")
            print(f"Total de {len(ids)} embeddings carregados, tamanho: {embeddings.shape}")

            # Verificar se o modelo é o mesmo
            if model_name != self.embedding_model_name:
                print(f"AVISO: O modelo usado para gerar os embeddings ({model_name}) é diferente do atual ({self.embedding_model_name})")

            return ids, embeddings, metadata

        except Exception as e:
            print(f"Erro ao carregar embeddings: {e}")
            return None, None, None

    def generate_and_save_embeddings(self, df, text_column='sentence', id_column='ProtocoloPedido',
                                   output_path=None, batch_size=128, include_metadata=False):
        """
        Gera embeddings para textos em um DataFrame e salva em pickle.

        Args:
            df: DataFrame com os textos
            text_column: Nome da coluna com os textos
            id_column: Nome da coluna com os IDs
            output_path: Caminho para salvar o arquivo pickle
            batch_size: Tamanho do lote para processamento
            include_metadata: Se True, inclui o DataFrame original no arquivo

        Returns:
            Tupla (ids, embeddings)
        """
        # Preparar textos e IDs
        texts = df[text_column].fillna('').astype(str).tolist()
        ids = df[id_column].tolist()

        # Definir caminho de saída se não fornecido
        if output_path is None:
            model_name_short = self.embedding_model_name.replace('/', '_')
            output_path = f"embeddings_{id_column}_{model_name_short}.pkl"

        # Gerar embeddings em lotes
        print(f"Gerando embeddings para {len(texts)} textos em lotes de {batch_size}...")
        start_time = time.time()

        all_embeddings = []

        for i in tqdm(range(0, len(texts), batch_size), desc="Gerando embeddings"):
            batch_texts = texts[i:i+batch_size]
            # Gera embeddings para esse lote
            batch_embeddings = self.embeddings.embed_documents(batch_texts)
            # Adiciona na lista final
            all_embeddings.extend(batch_embeddings)

        # Converter para array
        embeddings_array = np.array(all_embeddings, dtype='float32')

        print(f"Embeddings gerados em {time.time() - start_time:.2f} segundos")

        # Salvar em pickle
        metadata_to_include = df if include_metadata else None
        self.save_embeddings_to_pickle(ids, embeddings_array, output_path, metadata_to_include)

        return ids, embeddings_array

    def build_vectorstore_from_embeddings(self, ids, embeddings, store_type='pedidos', persist_directory=None):
        """
        Constrói um vectorstore a partir de embeddings pré-calculados.

        Args:
            ids: Lista de IDs (ProtocoloPedido ou IdRecurso)
            embeddings: Array numpy com os embeddings
            store_type: 'pedidos' ou 'recursos'
            persist_directory: Diretório para persistir o vectorstore (opcional)

        Returns:
            Vectorstore
        """
        from langchain.embeddings.base import Embeddings  # Importar a classe base Embeddings

        print(f"Construindo vectorstore para {len(ids)} {store_type}...")

        # Criar metadados com os IDs
        metadatas = []
        texts = []
        for id_val in ids:
            texts.append("")  # Texto vazio como placeholder
            if store_type == 'pedidos':
                metadata = {"ProtocoloPedido": str(id_val)}
            else:
                metadata = {"IdRecurso": str(id_val)}
            metadatas.append(metadata)

        # Criar uma classe wrapper que herda de Embeddings
        class PrecomputedEmbeddings(Embeddings):
            def __init__(self, embeddings_array, original_embedder):
                self.embeddings_array = embeddings_array
                self.original_embedder = original_embedder

            def embed_documents(self, _):
                # Retornar os embeddings pré-calculados, ignorando os textos
                return self.embeddings_array

            def embed_query(self, text):
                # Para consultas individuais, usar o embedder original
                return self.original_embedder.embed_query(text)

        # Guardar o embedder original
        original_embedder = self.embeddings

        # Substituir temporariamente com o wrapper
        try:
            self.embeddings = PrecomputedEmbeddings(embeddings, original_embedder)

            # Construir o vectorstore usando from_texts que aceita embeddings pré-calculados
            vectorstore = FAISS.from_texts(
                texts=texts,
                embedding=self.embeddings,
                metadatas=metadatas
            )

            # Salvar se especificado
            if persist_directory:
                os.makedirs(persist_directory, exist_ok=True)
                save_path = os.path.join(persist_directory, f"{store_type}.faiss")
                vectorstore.save_local(save_path)
                print(f"Vectorstore salvo em {save_path}")

            # Armazenar no objeto
            if store_type == 'pedidos':
                self.vectorstore_pedidos = vectorstore
            elif store_type == 'recursos':
                self.vectorstore_recursos = vectorstore

            return vectorstore

        finally:
            # Restaurar o embedder original
            self.embeddings = original_embedder

    def prepare_documents_from_dataframe(self, df, text_column='sentence', id_column='ProtocoloPedido'):
        """
        Prepara documentos a partir de um DataFrame para uso no LangChain.

        Args:
            df: DataFrame com os dados
            text_column: Nome da coluna contendo o texto
            id_column: Nome da coluna contendo o ID

        Returns:
            Lista de documentos LangChain
        """
        documents = []

        for _, row in df.iterrows():
            text = row[text_column]
            if pd.isna(text) or text == '':
                continue

            metadata = {col: row[col] for col in df.columns if col != text_column}
            # Converter valores numpy para Python nativos
            for key, value in metadata.items():
                if isinstance(value, np.integer):
                    metadata[key] = int(value)
                elif isinstance(value, np.floating):
                    metadata[key] = float(value)
                elif isinstance(value, np.ndarray):
                    metadata[key] = value.tolist()
                elif pd.isna(value):
                    metadata[key] = None

            doc = Document(page_content=text, metadata=metadata)
            documents.append(doc)

        return documents

    def create_vectorstore(self, documents, store_name="default", persist_directory=None):
        """
        Cria um vectorstore a partir de documentos.

        Args:
            documents: Lista de documentos LangChain
            store_name: Nome do armazenamento
            persist_directory: Diretório para persistir o vectorstore (opcional)

        Returns:
            Vectorstore
        """
        print(f"Criando vectorstore '{store_name}' com {len(documents)} documentos...")

        # Criar vectorstore FAISS
        vectorstore = FAISS.from_documents(
            documents=documents,
            embedding=self.embeddings
        )

        # Salvar se um diretório for especificado
        if persist_directory:
            os.makedirs(persist_directory, exist_ok=True)
            save_path = os.path.join(persist_directory, f"{store_name}.faiss")
            vectorstore.save_local(save_path)
            print(f"Vectorstore salvo em {save_path}")

        return vectorstore

    def build_pedidos_vectorstore(self, df_pedidos, persist_directory=None):
        """
        Constrói o vectorstore para pedidos.

        Args:
            df_pedidos: DataFrame com os pedidos
            persist_directory: Diretório para persistir (opcional)
        """
        # Preparar coluna 'sentence' se não existir
        if 'sentence' not in df_pedidos.columns:
            df_pedidos['sentence'] = (
                df_pedidos['ResumoSolicitacao'].fillna('')
                + ' <SEP> '
                + df_pedidos['DetalhamentoSolicitacao'].fillna('')
            )

        # Preparar documentos
        documents = self.prepare_documents_from_dataframe(df_pedidos)

        # Criar vectorstore
        self.vectorstore_pedidos = self.create_vectorstore(
            documents,
            store_name="pedidos",
            persist_directory=persist_directory
        )

        return self.vectorstore_pedidos

    def build_recursos_vectorstore(self, df_recursos, persist_directory=None):
        """
        Constrói o vectorstore para recursos.

        Args:
            df_recursos: DataFrame com os recursos
            persist_directory: Diretório para persistir (opcional)
        """
        # Preparar coluna 'sentence' se não existir
        if 'sentence' not in df_recursos.columns:
            df_recursos['sentence'] = (
                df_recursos['TipoRecurso'].fillna('')
                + ' <SEP> '
                + df_recursos['DescRecurso'].fillna('')
            )

        # Preparar documentos
        documents = self.prepare_documents_from_dataframe(
            df_recursos,
            id_column='IdRecurso'
        )

        # Criar vectorstore
        self.vectorstore_recursos = self.create_vectorstore(
            documents,
            store_name="recursos",
            persist_directory=persist_directory
        )

        return self.vectorstore_recursos

    def load_vectorstore(self, path, store_name="pedidos"):
        """
        Carrega um vectorstore salvo anteriormente.

        Args:
            path: Caminho base do diretório
            store_name: Nome do armazenamento (pedidos ou recursos)
        """
        store_path = os.path.join(path, f"{store_name}.faiss")

        if not os.path.exists(store_path):
            raise FileNotFoundError(f"Arquivo {store_path} não encontrado")

        print(f"Carregando vectorstore de {store_path}...")

        vectorstore = FAISS.load_local(
            store_path,
            self.embeddings
        )

        if store_name == "pedidos":
            self.vectorstore_pedidos = vectorstore
        elif store_name == "recursos":
            self.vectorstore_recursos = vectorstore

        return vectorstore

    def find_similar_pedidos(
        self,
        query_text=None,
        query_id=None,
        df_pedidos=None,
        k=10,
        filter_query=False
    ):
        """
        Encontra pedidos similares a partir de um texto ou ID.
        ...

        Retorna:
            Tupla (docs, results_df)
        """
        if self.vectorstore_pedidos is None:
            raise ValueError("Vectorstore de pedidos não inicializado")

        # --- normaliza o tipo do ID
        if query_id is not None:
            query_id = str(query_id)

        if query_id is not None and df_pedidos is not None:
            pedido = df_pedidos[df_pedidos['ProtocoloPedido'] == query_id]
            if pedido.empty:
                raise ValueError(f"Pedido com ID {query_id} não encontrado")

            query_text = pedido['sentence'].iloc[0]

        if query_text is None:
            raise ValueError("É necessário fornecer query_text ou query_id + df_pedidos")

        # Busca k vizinhos (+1 se vamos descartar o próprio)
        docs_with_scores = self.vectorstore_pedidos.similarity_search_with_score(
            query=query_text,
            k=k + 1 if filter_query else k
        )

        # Remove o próprio pedido se solicitado
        if filter_query and query_id is not None:
            docs_with_scores = [
                (doc, score) for doc, score in docs_with_scores
                if doc.metadata.get('ProtocoloPedido') != query_id   # já é string
            ]
            docs_with_scores = docs_with_scores[:k]  # garante no máximo k resultados

        # Empacota resultado em DataFrame
        results = []
        for doc, score in docs_with_scores:
            meta = doc.metadata.copy()
            meta['score'] = score
            meta['page_content'] = doc.page_content
            results.append(meta)

        results_df = pd.DataFrame(results)
        docs       = [doc for doc, _ in docs_with_scores]

        return docs, results_df

    def find_similar_recursos(self, query_text=None, query_id=None, df_recursos=None,
                           k=10, filter_query=False):
        """
        Encontra recursos similares a partir de um texto ou ID.

        Args:
            query_text: Texto da consulta (opcional)
            query_id: ID do recurso para consulta (opcional)
            df_recursos: DataFrame de recursos (necessário se query_id for usado)
            k: Número de resultados a retornar
            filter_query: Se True, remove o documento de consulta dos resultados

        Returns:
            Tupla (documentos, DataFrame de resultados)
        """
        if self.vectorstore_recursos is None:
            raise ValueError("Vectorstore de recursos não inicializado")

        if query_id is not None and df_recursos is not None:
            # Buscar texto do recurso pelo ID
            recurso = df_recursos[df_recursos['IdRecurso'] == query_id]
            if len(recurso) == 0:
                raise ValueError(f"Recurso com ID {query_id} não encontrado")

            query_text = recurso['sentence'].iloc[0]

        if query_text is None:
            raise ValueError("É necessário fornecer query_text ou query_id + df_recursos")

        # Realizar busca
        docs_with_scores = self.vectorstore_recursos.similarity_search_with_score(
            query=query_text,
            k=k+1 if filter_query else k  # Pedir um extra se vamos filtrar
        )

        # Filtrar documento de consulta se necessário
        if filter_query and query_id is not None:
            docs_with_scores = [
                (doc, score) for doc, score in docs_with_scores
                if doc.metadata.get('IdRecurso') != query_id
            ]
            # Limitar aos k primeiros após filtro
            docs_with_scores = docs_with_scores[:k]

        # Criar DataFrame com resultados
        results = []
        for doc, score in docs_with_scores:
            metadata = doc.metadata.copy()
            metadata['score'] = score
            metadata['page_content'] = doc.page_content
            results.append(metadata)

        results_df = pd.DataFrame(results)

        # Separar documentos para retorno
        docs = [doc for doc, _ in docs_with_scores]

        return docs, results_df

    def batch_similar_search(self, query_ids, df, vectorstore_type='pedidos', k=10):
        """
        Realiza busca em lote para vários IDs.

        Args:
            query_ids: Lista de IDs para consulta
            df: DataFrame com os dados
            vectorstore_type: 'pedidos' ou 'recursos'
            k: Número de resultados por consulta

        Returns:
            DataFrame com resultados consolidados
        """
        all_results = []

        for qid in tqdm(query_ids, desc=f"Buscando similares para {len(query_ids)} {vectorstore_type}"):
            try:
                if vectorstore_type == 'pedidos':
                    _, results = self.find_similar_pedidos(
                        query_id=qid,
                        df_pedidos=df,
                        k=k,
                        filter_query=True
                    )
                    id_field = 'ProtocoloPedido'
                else:
                    _, results = self.find_similar_recursos(
                        query_id=qid,
                        df_recursos=df,
                        k=k,
                        filter_query=True
                    )
                    id_field = 'IdRecurso'

                # Adicionar ID da consulta
                results['query_id'] = qid

                # Filtrar colunas relevantes
                if len(results) > 0:
                    results = results[[
                        'query_id',
                        id_field,
                        'score',
                        'page_content'
                    ]]

                    all_results.append(results)
            except Exception as e:
                print(f"Erro ao processar {vectorstore_type} {qid}: {e}")

        # Consolidar resultados
        if all_results:
            return pd.concat(all_results, ignore_index=True)
        else:
            return pd.DataFrame()

    def evaluate_with_annotated_data(self, df_annotated, df_pedidos, k_values=[1, 5, 10, 20, 50, 100]):
        """
        Avalia o desempenho do pipeline usando dados anotados.

        Args:
            df_annotated: DataFrame com pares de pedidos anotados (colunas: 'NUP', 'NUP_semelhante')
            df_pedidos: DataFrame com os pedidos
            k_values: Lista de valores de k para avaliação

        Returns:
            DataFrame com métricas de avaliação
        """
        # Verificar se vectorstore de pedidos está inicializado
        if self.vectorstore_pedidos is None:
            raise ValueError("Vectorstore de pedidos não inicializado")

        results = []
        unique_queries = df_annotated['NUP'].unique()

        print(f"Avaliando {len(unique_queries)} consultas com {len(df_annotated)} pares anotados")

        # Para cada valor de k
        for k in k_values:
            hits = 0
            total_pairs = 0

            # Para cada consulta
            for query_nup in tqdm(unique_queries, desc=f"Avaliando com k={k}"):
                # Obter NUPs similares anotados
                similar_nups = df_annotated[df_annotated['NUP'] == query_nup]['NUP_semelhante'].tolist()

                # Buscar similares com o modelo
                _, results_df = self.find_similar_pedidos(
                    query_id=query_nup,
                    df_pedidos=df_pedidos,
                    k=k,
                    filter_query=True
                )

                # Verificar hits
                predicted_nups = results_df['ProtocoloPedido'].tolist()
                for nup in similar_nups:
                    total_pairs += 1
                    if nup in predicted_nups:
                        hits += 1

            # Calcular métricas
            recall = hits / total_pairs if total_pairs > 0 else 0
            results.append({
                'k': k,
                'hits': hits,
                'total_pairs': total_pairs,
                'recall': recall
            })

        return pd.DataFrame(results)
```

```python
def busca_pedidos_similares_langchain(nup, df_pedidos, pipeline, k=10):
    """
    Retorna os top-k pedidos similares ao NUP informado usando LangChain.

    Args:
        nup: Número do protocolo do pedido
        df_pedidos: DataFrame com os pedidos
        pipeline: Pipeline de recomendação LangChain
        k: Número de resultados a retornar

    Returns:
        Tupla (docs, DataFrame com resultados)
    """
    # Usar o método find_similar_pedidos do pipeline
    docs, results_df = pipeline.find_similar_pedidos(
        query_id=nup,
        df_pedidos=df_pedidos,
        k=k,
        filter_query=True  # Já remove o próprio pedido
    )

    return docs, results_df
```

```python
def simjoin_pedidos_langchain(protocolos, df_pedidos, pipeline, k=10):
    """
    Para cada protocolo em 'protocolos', busca os k similares usando LangChain
    e devolve um DataFrame [pedido, pedido_similar, score].
    """
    # garante que todos os IDs são string
    protocolos_str = [str(p) for p in protocolos]

    # busca em lote
    batch_results = pipeline.batch_similar_search(
        query_ids=protocolos_str,
        df=df_pedidos,
        vectorstore_type="pedidos",
        k=k
    )

    # se não houve resultados, devolve DF vazio com as colunas corretas
    if batch_results.empty:
        return pd.DataFrame(columns=["pedido", "pedido_similar", "score"])

    # renomeia para o formato esperado
    recomendacoes = (batch_results
                     .rename(columns={"query_id": "pedido",
                                      "ProtocoloPedido": "pedido_similar"})
                     [["pedido", "pedido_similar", "score"]])

    return recomendacoes
```

```python
# DEPRECATED: Versão antiga da função calcula_metricas_at_k (mantemos aqui só por referência)
from sklearn.metrics import recall_score

def calcula_metricas_at_k_deprecated(recomendacoes, rotulos_verdadeiros, k):
    # recomendacoes: DataFrame com colunas ['pedido', 'pedido_similar', 'score']
    # rotulos_verdadeiros: DataFrame com colunas ['NUP', 'NUP_semelhante']
    # k: número de top recomendações a considerar

    recalls = []
    recalls_macro = []
    precisions = []
    f1_scores = []

    for nup_query in recomendacoes['pedido'].unique():
        # Obter os verdadeiros similares para o pedido
        verdadeiros_similares = set(rotulos_verdadeiros[rotulos_verdadeiros['NUP'] == nup_query]['NUP_semelhante'])
        # Obter os top-k similares recomendados
        top_k_recomendados = set(recomendacoes[recomendacoes['pedido'] == nup_query].nlargest(k, 'score')['pedido_similar'])

        # TP, FP, FN
        verdadeiros_positivos = len(verdadeiros_similares.intersection(top_k_recomendados))
        falsos_positivos = len(top_k_recomendados) - verdadeiros_positivos
        falsos_negativos = len(verdadeiros_similares) - verdadeiros_positivos

        # Conjunto total
        conjunto_total_itens = set()
        conjunto_total_itens.update(verdadeiros_similares)
        conjunto_total_itens.update(top_k_recomendados)

        # Preparar vetores para sklearn
        y_true = [1 if item in verdadeiros_similares else 0 for item in conjunto_total_itens]
        y_pred = [1 if item in top_k_recomendados else 0 for item in conjunto_total_itens]

        # Cálculo manual de recall e precisão
        recall = (verdadeiros_positivos / (verdadeiros_positivos + falsos_negativos)) if (verdadeiros_positivos + falsos_negativos) > 0 else 0
        precisao = (verdadeiros_positivos / k) if k > 0 else 0

        # F1 manual
        f1_score_val = 2 * (precisao * recall) / (precisao + recall) if (precisao + recall) > 0 else 0

        # Recall macro via sklearn
        recall_macro = recall_score(y_true, y_pred, average='macro', zero_division=0)

        recalls_macro.append(recall_macro)
        recalls.append(recall)
        precisions.append(precisao)
        f1_scores.append(f1_score_val)

    recall_medio_at_k = sum(recalls) / len(recalls) if recalls else 0
    precisao_media_at_k = sum(precisions) / len(precisions) if precisions else 0
    f1_score_medio_at_k = sum(f1_scores) / len(f1_scores) if f1_scores else 0
    recall_medio_macro_at_k = sum(recalls_macro) / len(recalls_macro) if recalls_macro else 0

    return recall_medio_at_k, recall_medio_macro_at_k, precisao_media_at_k, f1_score_medio_at_k
```

```python
from sklearn.metrics import recall_score, precision_score, f1_score

def calcula_metricas_at_k(recomendacoes, rotulos_verdadeiros, k):
    """
    recomendacoes: DataFrame ['pedido', 'pedido_similar', 'score']
    rotulos_verdadeiros: DataFrame ['NUP', 'NUP_semelhante']
    """
    recalls, recalls_macro, precisions, f1_scores = [], [], [], []

    for nup_query in recomendacoes['pedido'].unique():
        nup_query = str(nup_query)                                   # ← garante string

        # ground-truth
        verdadeiros_similares = set(
            rotulos_verdadeiros[rotulos_verdadeiros['NUP'] == nup_query]['NUP_semelhante']
            .astype(str)
        )

        # top-k recomendados (menor distância → mais perto)
        top_k_recomendados = set(
            recomendacoes[recomendacoes['pedido'] == nup_query]
              .nsmallest(k, 'score')['pedido_similar']
              .astype(str)
        )

        if not verdadeiros_similares:
            continue

        universo = verdadeiros_similares | top_k_recomendados
        y_true = [1 if x in verdadeiros_similares else 0 for x in universo]
        y_pred = [1 if x in top_k_recomendados   else 0 for x in universo]

        recalls.append(recall_score(y_true, y_pred, zero_division=0))
        recalls_macro.append(recall_score(y_true, y_pred, average='macro', zero_division=0))
        precisions.append(precision_score(y_true, y_pred, zero_division=0))
        f1_scores.append(f1_score(y_true, y_pred, zero_division=0))

    return (
        sum(recalls)        / len(recalls)        if recalls else 0,
        sum(recalls_macro)  / len(recalls_macro)  if recalls_macro else 0,
        sum(precisions)     / len(precisions)     if precisions else 0,
        sum(f1_scores)      / len(f1_scores)      if f1_scores else 0,
    )
```

```python
from typing import Set

def calcula_map(recomendacoes: pd.DataFrame,
                rotulos_verdadeiros: pd.DataFrame,
                k_max: int = 100) -> float:
    """
    Cálculo do Mean Average Precision (mAP).

    recomendacoes      DF ['pedido', 'pedido_similar', 'score']
    rotulos_verdadeiros DF ['NUP', 'NUP_semelhante']
    k_max              até quantos vizinhos considerar
    """
    if recomendacoes.empty:
        return 0.0

    rotulos_verdadeiros = rotulos_verdadeiros.astype(str)
    recomendacoes[['pedido', 'pedido_similar']] = (
        recomendacoes[['pedido', 'pedido_similar']].astype(str)
    )

    average_precisions = []

    for nup_query in recomendacoes['pedido'].unique():
        gt: Set[str] = set(
            rotulos_verdadeiros.loc[rotulos_verdadeiros['NUP'] == nup_query,
                                    'NUP_semelhante']
        )
        if not gt:
            continue

        # ordena do MAIS parecido (score menor) para o menos
        recomendados = (
            recomendacoes.loc[recomendacoes['pedido'] == nup_query]
                         .sort_values('score', ascending=True)
                         .head(k_max)['pedido_similar']
        )

        precisions = []
        hits = 0

        for rank, rec in enumerate(recomendados, start=1):
            if rec in gt:
                hits += 1
                precisions.append(hits / rank)

        # média de precisões para este pedido
        ap = sum(precisions) / min(len(gt), k_max)
        average_precisions.append(ap)

    return sum(average_precisions) / len(average_precisions) if average_precisions else 0.0
```

```python
def calcula_mrr(recomendacoes: pd.DataFrame,
                rotulos_verdadeiros: pd.DataFrame) -> float:
    """
    Mean Reciprocal Rank (MRR).

    recomendacoes      DF ['pedido', 'pedido_similar', 'score']
    rotulos_verdadeiros DF ['NUP', 'NUP_semelhante']
    """
    if recomendacoes.empty:
        return 0.0

    # normaliza tipo string
    rotulos_verdadeiros = rotulos_verdadeiros.astype(str)
    recomendacoes[['pedido', 'pedido_similar']] = (
        recomendacoes[['pedido', 'pedido_similar']].astype(str)
    )

    reciprocal_ranks = []

    for nup_query in recomendacoes['pedido'].unique():
        gt = set(
            rotulos_verdadeiros.loc[rotulos_verdadeiros['NUP'] == nup_query,
                                    'NUP_semelhante']
        )
        if not gt:
            continue

        # ordena do mais parecido (menor score) para o menos
        recomendados = (
            recomendacoes.loc[recomendacoes['pedido'] == nup_query]
                         .sort_values('score', ascending=True)['pedido_similar']
        )

        # posição do primeiro relevante
        rank = next((idx + 1
                     for idx, rec in enumerate(recomendados)
                     if rec in gt), None)

        if rank:
            reciprocal_ranks.append(1 / rank)

    return sum(reciprocal_ranks) / len(reciprocal_ranks) if reciprocal_ranks else 0.0
```

```python
import numpy as np

def calcula_ndcg(recomendacoes: pd.DataFrame,
                 rotulos_verdadeiros: pd.DataFrame,
                 k: int = 10) -> float:
    """
    NDCG@k médio sobre todas as consultas.
    """
    if recomendacoes.empty:
        return 0.0

    # garante string
    rotulos_verdadeiros = rotulos_verdadeiros.astype(str)
    recomendacoes[['pedido', 'pedido_similar']] = (
        recomendacoes[['pedido', 'pedido_similar']].astype(str)
    )

    ndcg_scores = []

    for nup_query in recomendacoes['pedido'].unique():
        gt = set(
            rotulos_verdadeiros.loc[rotulos_verdadeiros['NUP'] == nup_query,
                                    'NUP_semelhante']
        )
        if not gt:
            continue

        # top-k pelo MENOR score (mais parecido)
        top_k_recomendados = (
            recomendacoes.loc[recomendacoes['pedido'] == nup_query]
                         .nsmallest(k, 'score')['pedido_similar']
        )

        # -------- DCG@k --------
        dcg = 0.0
        for rank, rec in enumerate(top_k_recomendados, start=1):
            if rec in gt:
                dcg += 1 / np.log2(rank + 1)     # rank começa em 1

        # -------- IDCG@k --------
        m = min(len(gt), k)                     # nº relev. possíveis no top-k
        idcg = sum(1 / np.log2(r + 1) for r in range(1, m + 1))

        ndcg_scores.append(dcg / idcg if idcg else 0.0)

    return sum(ndcg_scores) / len(ndcg_scores) if ndcg_scores else 0.0
```

```python
# Função para obter os recursos de um NUP
def get_recursos(nup, df_recursos):
    """
    Filtra os recursos associados a um NUP.

    Args
    ----
    nup : str | int
        Número de protocolo do pedido.
    df_recursos : pd.DataFrame
        DataFrame contendo, obrigatoriamente, a coluna 'ProtocoloPedido'.

    Returns
    -------
    pd.DataFrame | str
        DataFrame com os recursos encontrados ou 'null' se nenhum registro.
    """
    nup_str = str(nup).strip()                        # garante string

    # se ainda não converteu antes:
    if df_recursos['ProtocoloPedido'].dtype != "object":
        df_recursos = df_recursos.copy()
        df_recursos['ProtocoloPedido'] = df_recursos['ProtocoloPedido'].astype(str)

    filtrado = df_recursos.loc[df_recursos['ProtocoloPedido'] == nup_str]

    return filtrado if not filtrado.empty else "null"
```

```python
def busca_recursos_similares(id_recurso, df_recursos, pipeline, k=10):
    """
    Retorna os k recursos mais similares ao recurso `id_recurso`.

    Parameters
    ----------
    id_recurso : str | int
        Identificador do recurso de consulta.
    df_recursos : pd.DataFrame
        DataFrame completo de recursos.
    pipeline : CGURecommendationPipeline
        Instância já inicializada com o vector-store de recursos.
    k : int, default 10
        Quantidade de vizinhos a retornar.

    Returns
    -------
    docs : list[Document]
        Objetos Document retornados pelo LangChain.
    results_df : pd.DataFrame
        DataFrame com colunas metadata + score + page_content.
    """
    id_recurso_str = str(id_recurso).strip()            # garante string

    docs, results_df = pipeline.find_similar_recursos(
        query_id=id_recurso_str,
        df_recursos=df_recursos,
        k=k,                   # NÃO soma +1 aqui; a função já faz isso
        filter_query=True       # remove o próprio recurso
    )

    return docs, results_df
```

```python
def simjoin_recursos(ids_recursos, df_recursos, pipeline, k=10):
    """
    Para cada recurso em `ids_recursos`, busca os k vizinhos mais similares.

    Returns
    -------
    pares_recursos_similares : list[[recurso, recurso_similar, score]]
    recomendacoes_recursos   : pd.DataFrame  (colunas: recurso, recurso_similar, score)
    """
    # garante tipo string
    ids_str = [str(rid) for rid in ids_recursos]

    # busca em lote
    batch_results = pipeline.batch_similar_search(
        query_ids=ids_str,
        df=df_recursos,
        vectorstore_type="recursos",
        k=k
    )

    # se não houve retorno, devolve estruturas vazias no formato correto
    if batch_results.empty:
        empty_df = pd.DataFrame(columns=["recurso", "recurso_similar", "score"])
        return [], empty_df

    # renomeia e mantém só as colunas relevantes
    recomendacoes_recursos = (
        batch_results
          .rename(columns={"query_id": "recurso", "IdRecurso": "recurso_similar"})
          [["recurso", "recurso_similar", "score"]]
    )

    # DataFrame → lista de listas
    pares_recursos_similares = recomendacoes_recursos.values.tolist()

    return pares_recursos_similares, recomendacoes_recursos
```

```python
import time

def exemplo_busca_recursos(pipeline, dt_recursos, id_recurso_exemplo=None, k=10):
    """
    Executa um exemplo de busca de recursos similares.
    """
    # 1. Seleciona o ID de exemplo e normaliza para string
    if id_recurso_exemplo is None:
        id_recurso_exemplo = dt_recursos['IdRecurso'].iloc[0]

    id_recurso_exemplo = str(id_recurso_exemplo).strip()

    print(f"\nRealizando busca de recursos similares para ID {id_recurso_exemplo} (k={k})")

    recurso_exemplo = dt_recursos.loc[dt_recursos['IdRecurso'] == id_recurso_exemplo]
    if not recurso_exemplo.empty:
        desc = recurso_exemplo['DescRecurso'].iloc[0]
        print("\nDescrição do recurso exemplo:")
        print(desc[:200], "...\n")

    # 2. Busca similares
    start = time.time()
    docs, results_df = busca_recursos_similares(
        id_recurso_exemplo,
        dt_recursos,
        pipeline,
        k=k
    )
    print(f"Tempo para processamento: {time.time() - start:.2f}s")

    if results_df.empty:
        print("Nenhum recurso similar encontrado.")
        return results_df

    print(f"Quantidade de recursos similares encontrados: {len(results_df)}\n")

    # 3. Exibe resultados
    print("Resultados da busca:")
    for rank, row in results_df.iterrows():
        rec_id  = row['IdRecurso']
        score   = row['score']
        desc_re = dt_recursos.loc[dt_recursos['IdRecurso'] == rec_id, 'DescRecurso']
        print(f"{rank+1:>2}. ID: {rec_id} | Score: {score:.4f}")
        if not desc_re.empty:
            print(f"    Descrição: {desc_re.iloc[0][:100]}...\n")

    return results_df
```

```python
def avalia_recursos_similares(pipeline,
                              dt_recursos,
                              df_anotado_recursos=None,
                              k_values=[1, 5, 10, 20, 50]):
    """
    Avalia a busca de recursos similares.

    Se `df_anotado_recursos` (ground-truth) não for passado,
    devolve apenas estatísticas dos scores.
    """
    # 1. Amostra de recursos
    unique_ids = dt_recursos['IdRecurso'].astype(str).unique()
    sample_size = min(50, len(unique_ids))
    recursos_amostra = np.random.default_rng(42).choice(unique_ids,
                                                        size=sample_size,
                                                        replace=False).tolist()

    print(f"Avaliando {sample_size} recursos (k máx = {max(k_values)}) …")

    # 2. Busca em lote
    t0 = time.time()
    batch_results = pipeline.batch_similar_search(
        query_ids=[str(rid) for rid in recursos_amostra],   # garante string
        df=dt_recursos,
        vectorstore_type="recursos",
        k=max(k_values)
    )
    took = time.time() - t0

    print(f"Tempo total: {took:.2f}s  |  médio: {took/sample_size:.4f}s por recurso")

    if batch_results.empty:
        print("⚠️  Nenhum vizinho retornado.")
        return pd.DataFrame()     # devolve DF vazio

    # 3. Estatísticas básicas dos scores
    print("\nEstatísticas dos scores:")
    print(batch_results['score'].describe()[['mean', 'min', 'max', 'std']])

    # 4. Métricas se houver dados anotados
    if df_anotado_recursos is None:
        return batch_results

    # rename p/ compatibilidade com funções de métrica
    recs = (batch_results
            .rename(columns={"query_id": "pedido",
                             "IdRecurso": "pedido_similar"})
            [["pedido", "pedido_similar", "score"]])

    # garante string no ground-truth
    df_anotado_recursos = df_anotado_recursos.astype(str)

    resultados = []
    for k in k_values:
        recall_k, recall_macro_k, prec_k, f1_k = calcula_metricas_at_k(
            recs, df_anotado_recursos, k
        )
        map_k   = calcula_map (recs, df_anotado_recursos, k)
        ndcg_k  = calcula_ndcg(recs, df_anotado_recursos, k)
        mrr_k   = calcula_mrr (recs, df_anotado_recursos) if k == max(k_values) else None

        resultados.append({
            "k": k,
            "recall": recall_k,
            "precision": prec_k,
            "f1_score": f1_k,
            "map": map_k,
            "mrr": mrr_k,
            "ndcg": ndcg_k
        })

    return pd.DataFrame(resultados)
```

```python
# Criar o prompt COM few-shot

# Dados do recurso de entrada
tipo_recurso = recurso_exemplo['TipoRecurso']
descricao_recurso = recurso_exemplo['DescRecurso']
orgao_pedido = recurso_exemplo['OrgaoPedido']

# Mensagem de sistema (instrução few-shot)
messages_fs = [
    {"role": "system", "content": instruction_fs}
]

# Gerar 2 exemplos (few-shot)
fs = generate_few_shot(2)
messages_fs.extend(fs)

# Mensagem do usuário com o recurso que estamos analisando
entrada_recurso = {
    "role": "user",
    "content": (
        f"<tipo_recurso> {tipo_recurso} </tipo_recurso> "
        f"<recurso> {descricao_recurso} </recurso> "
        f"<orgao> {orgao_pedido} </orgao>"
    )
}
messages_fs.append(entrada_recurso)
```

```python
def build_chat_prompt(messages):
    """
    Converte uma lista de dicionários com 'role' e 'content' em uma única string
    que sirva de prompt para o pipeline.
    Exemplo de formatação:
      "System: <conteudo>\nUser: <conteudo>\nAssistant: <conteudo>\n..."
    """
    final_text = ""
    for msg in messages:
        role = msg["role"]
        content = msg["content"]

        if role == "system":
            final_text += f"System: {content}\n\n"
        elif role == "assistant":
            final_text += f"Assistant: {content}\n\n"
        else:  # "user"
            final_text += f"User: {content}\n\n"

    return final_text
```

```python
# Inferência no LLM para gerar resposta (few-shot)
start = time.time()

# Montar o prompt concatenado
prompt_fs = build_chat_prompt(messages_fs)

# Chamar o pipeline
outputs_fs = pipe(prompt_fs,
                  do_sample=True,
                  temperature=0.7,
                  top_k=20,
                  top_p=0.95)

end = time.time()
print(f"Tempo para gerar a resposta (few-shot): {end - start:.2f}s")

# O pipeline retorna uma lista de dicionários.
# Em text-generation, costuma ser algo como outputs_fs[0]['generated_text']
resposta_fs = outputs_fs[0]["generated_text"]
```
